```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# os.getcwd()
# file_dir = r'C:\Users\user\Documents\Oren\Data Science_NAYA_02-2023\datasets_partial'
# diam_fname = 'diamonds.csv'
# full_path = file_dir + '\\' + diam_fname
# full_path

diam_path = r'C:\Users\user\Documents\Oren\Data Science_NAYA_02-2023\datasets_partial\diamonds.csv'
data = pd.read_csv(diam_path)
data.set_index('#', inplace = True)
data = data[['carat','price']]
```

```python
# plotting to examine whether relation between 'carat' and 'price'
# can be described by a straight line (Answer: no...)

x1 = data['carat']
y1 = data['price']

fig = plt.figure(figsize = (4,4))
plt.xlim((0,3))
plt.ylim((0,20000))
ax = fig.gca()
ax.plot(x1,y1, '.r', markersize = 1)

ax.set_title('diamonds: carat vs. price')
ax.set_xlabel('carat')
ax.set_ylabel('price')

#find line of best fit
a, b = np.polyfit(x1, y1, 1)
#add line of best fit to plot
plt.plot(x1, a*x1+b)
# Creating a model-based y-variable calculation
mod_ln_price = a*x1+b
# data['price_test'] = mod_ln_price.apply(lambda x: np.exp(x))
```

```python
# calculating natural-log('carat'), natural-log('price')
data['natural_log(price)'] = np.log(data['price'])
data['natural_log(carat)'] = np.log(data['carat'])
```

```python
# Approach #2:
# Plotting to test whether log(carat) vs. log(price) could be described by a straight line
# Answer (Better...)
x2 = data['natural_log(carat)']
y2 = data['natural_log(price)']

fig = plt.figure(figsize = (4,4))
plt.ylim((5,11))
ax = fig.gca()
ax.plot(x2,y2, '.r', markersize = 1)

ax.set_title('diamonds: natural-log(carat) vs. natural-log(price)')
ax.set_xlabel('natural-log(carat)')
ax.set_ylabel('natural-log(price)')

#find line of best fit
a2, b2 = np.polyfit(x2, y2, 1)
#add line of best fit to plot
plt.plot(x2, a2*x2+b2)
# Creating a model-based y-variable calculation
mod_ln_price2 = a2*x2+b2
data['price_test2'] = mod_ln_price2.apply(lambda x2: math.exp(x2)).round(2)
```

```python
# Calculating how far the actual price from model price are, based on approach 2:
data['ratio_test2'] = data['price']/data['price_test2']
data['ratio_test2'].describe()
```
